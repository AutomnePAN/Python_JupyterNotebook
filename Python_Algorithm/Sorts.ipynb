## Shell Sort

```python
def shell_sort(collection):
    gap = len(collection)//2
    while(gap >= 1):
        i = gap;
#         print(gap)
        while( i < len(collection) ):
            j = i
            temp = collection[i];
#             print("temp:",temp)
#             print("current: ",collection[j-gap])
            while j >= gap and collection[j - gap] > temp:
                collection[j] = collection[j - gap] 
                j = j - gap;
            collection[j] = temp
            i += 1
        gap = gap//2
    return collection
```

```python
Test = [random.random() for r in range(100)]
shell_sort(Test)
```

## Insertion Sort

```python
def insertion_sort(collection):
    for i in range (len(collection)):
        temp = collection[i]
        j = i
        while j>=1 and collection[j-1] > temp:
            collection[j] = collection[j-1]
            j -= 1
        collection[j] = temp
    return collection;
```

```python
Test = [random.random() for r in range(100)]
insertion_sort(Test)
```

## Selection Sort

```python
def selection_sort(collection):
    for i in range(len(collection)):
        min = collection[i]
        for j in range(i+1, len(collection)):
            if collection[j] < min:
                collection[j], min = min, collection[j]
        collection[i] = min
    return collection
    
```

```python
Test = [random.random() for r in range(100)]
selection_sort(Test)
```

## Bucket Sort

```python
def bucket_sort(collection, bucket_size = 5):
    if (len(collection) == 0):
        return collection
    min_val = collection[0]
    max_val = collection[0]
    
    for i in range(len(collection)):
        if collection[i] < min_val:
            min_val = collection[i]
        elif collection[i] > max_val:
            max_val = collection[i]
    
    bucket_num = math.floor((max_val - min_val)/bucket_size) + 1
    buckets = []
    for i in range(bucket_num):
        buckets.append([])
    for i in range(len(collection)):
        buckets[math.floor((collection[i] - min_val)/bucket_size)].append(collection[i])
        
    Sorted_collection = []
    for i in range(0, len(buckets)):
        insertion_sort(buckets[i])
        for j in range(0, len(buckets[i])):
            Sorted_collection.append(buckets[i][j])
            
    return Sorted_collection
```

## Quick Sort

```python
def quick_sort(collection):
    if len(collection) < 1:
        return collection
    pivot = collection[0]
    greater = [element for element in collection[1:] if element >= pivot]
    lesser = [element for element in collection[1:] if element < pivot]
    return quick_sort(lesser)+[pivot]+quick_sort(greater)
```

```python
Test = [random.randint(0,3000) for i in range(100000)]
print(Test)
quick_sort(Test)
```

## Heap Sort

```python
def heapify(collection, index, heap_size):
    largest = index
    left_index = 2*index + 1;
    right_index = 2*index + 2
    if left_index < heap_size and collection[left_index] > collection[largest]:
        largest = left_index
    if right_index < heap_size and collection[right_index] > collection[largest]:
        largest = right_index
    if largest != index:
        collection[largest], collection[index] = collection[index], collection[largest]
        heapify(collection, largest, heap_size)
```

```python
def heap_sort(collection):
    n = len(collection)
    for i in range(n // 2 - 1, -1,-1):
        heapify(collection,i,n)
    for i in range(n-1, -1, -1):
        collection[0], collection[i] = collection[i], collection[0]
        heapify(collection,0, i)
    return collection
```
