```python
class Solution:
    """
    @param m: An integer m denotes the size of a backpack
    @param A: Given n items with size A[i]
    @param V: Given n items with value V[i]
    @return: The maximum value
    """
    def backPackII(self, m, A, V):
        # write your code here
        n = len(A);
        dp = [ 0 for i in range(m+1)]
        for i in range(n):
            for j in range(m, A[i] - 1, -1):
                dp[j] = max(dp[j], dp[j-A[i]] + V[i]);
        return dp[m]
```

```python
"""
Definition of ListNode
class ListNode(object):

    def __init__(self, val, next=None):
        self.val = val
        self.next = next
"""
class Solution:
    """
    @param hashTable: A list of The first node of linked list
    @return: A list of The first node of linked list which have twice size
    """
    def rehashing(self, hashTable):
        # write your code here
        ori_cap = len(hashTable);
        new_cap = 2 * ori_cap;
        res_hashtable = [None for i in range(new_cap)];
    
        
        for start_node in hashTable:
            node = start_node;
            while node != None:
                key = node.val % new_cap;
                if res_hashtable[key] == None:
                    res_hashtable[key] = ListNode(node.val);
                else:
                    node_f = res_hashtable[key]
                    while node_f.next:
                        node_f = node_f.next;
                    node_f.next = ListNode(node.val);
                node = node.next;
        return res_hashtable;
```
