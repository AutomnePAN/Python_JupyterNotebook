```python
class Solution:
    """
    @param: A: An integer array
    @param: target: An integer
    @return: An integer
    """
    def MinAdjustmentCost(self, A, target):
        # write your code here
        f = [[ 0 for j in range(100)] for i in range(len(A)) ]
        for i in range(1,len(A)):
            for j in range(100):
                
                candidat = []
                for k in range(j-target, j+target+1):
                    if k >= 0 and k < 100:
                        candidat.append(k);
                print(candidat);
                f[i][j] = min([f[i-1][k] for k in candidat]) + abs(A[i-1] - j);
        return min(f[len(A) - 1])
```

```python
class Solution:
    """
    @param m: An integer m denotes the size of a backpack
    @param A: Given n items with size A[i]
    @return: The maximum size
    """
    def backPack(self, m, A):
        # write your code here
        n = len(A)
        dp = [[0 for j in range(m+1)] for _ in range(n+1)]
        
         for i in range(1, n+1):
            for j in range(1, m+1):
                dp[i][j] = dp[i-1][j]
                if j - A[i-1] >= 0:
                    dp[i][j] = max(dp[i][j], dp[i-1][j - A[i-1]] + A[i-1])
    
        return dp[-1][-1]
```

```python
"""
Definition of TreeNode:
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left, self.right = None, None
"""

class Solution:
    """
    @param root: The root of binary tree.
    @return: True if the binary tree is BST, or false
    """
    def isValidBST(self, root):
        # write your code here
        valid, _, _ = self.helper(root);
        return valid;
        
    def helper(self, root):
        if root == None:
            return True, 0, 0;
        
        
        if root.left:
            left = self.helper(root.left);
        if root.right:
            right = self.helper(root.right);
        if not root.left:
            return right[0] and root.val < right[1], min(root.val, right[1]), max(root.val, right[2])
        if not root.right:
            return left[0] and root.val > left[2], min(root.val, left[1]), max(root.val, left[2])
        return left[0] and right[0] and left[2] < root.val < right[1], min(left[1], root.val, right[1]) , max(left[2], root.val, right[2])
```
